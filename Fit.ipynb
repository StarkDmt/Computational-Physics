```python
import numpy as np
from scipy.optimize import curve_fit

a=0.5
b=1.0
c=2.0
s=0.05
left=-3.0
right=3.0
density=20
amplitude=2.0

def f(x,a,b,c): 
    return c*(1 + ( (x - a) / b ) ** 2 ) ** (-1)

points = np.linspace(left,right,density)
np.random.seed(11037)
noise = np.random.normal(scale = s, size = points.size)
values = [f(x,a,b,c) for x in points]
data = values + amplitude * noise   

fit,matrix = curve_fit(f,points,data,[1,1,1])
print(fit)

import matplotlib.pyplot as plt

plt.plot(points, f(points,a,b,c), '--',label = 'The Truest of Truths')
plt.plot(points, data, 'o',label = 'Noisy Truth')
plt.plot(points, f(points,*fit), '-', label = 'Close to Truth')
plt.legend (loc ='best' )
plt.show()
```
