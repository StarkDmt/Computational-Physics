```python
import matplotlib.pyplot as plt
def f(x):
    return x**11
```

```python
def derivative(x):
  return 11*x**10

def left_approximation(x,h):
  return (f(x + h) - f(x))/h

def centre_approximation(x,h):
  return (f(x + h) - f (x - h))/(2 * h)
```

Построим графики разностей аналитической и численно вычисленных производных в логарифмическом масштабе.

```python
x=2.0

xset=[10 ** (k/10) for k in range (-200, -10)]
yset=[left_approximation(x,e) - derivative(x) for e in xset]
```

Для левостороннего приближения:

```python
plt.loglog (xset,yset, 'o-')
plt.grid(True)
plt.show()
```

```python
yset=[centre_approximation(x,e) - derivative(x) for e in xset]
```

Для центрального приближения:

```python
plt.loglog (xset,yset, 'o-')
plt.grid(True)
plt.show()
```

Хаотичное поведение левых веток обоих графиков вызвано накоплением ошибок округления. Видно, что вторая схема позволяет получить на 2 порядка более точный результат. Тангенс угла наклона правой ветки для первой схемы равен 1 (линейная зависимость от h), для второй - 2 (квадратичная зависимость от h).
