```python
import math
def lagrange_i(i,xset,point):
    res = 1.0
    for k in range(len(xset)):
        if(k != i):
            res*=(point - xset[k])/(xset[i]-xset[k])
    return res

def lagrange(xset,yset,point):
    return sum(yset[i]*lagrange_i(i,xset,point) for i in range(len(xset)))

def f(x):
    return 1/(1+25*x**2)
```

```python
mesh_precision=9
chebysh_precision=9

regular_mesh_x = [(2.0 / mesh_precision * i - 1) for i in range(mesh_precision)]
regular_mesh_y = [f(regular_mesh_x[i]) for i in range (len(regular_mesh_x))]

chebyshev_mesh_x = [(math.cos((2 * i + 1) / (2*chebysh_precision) * math.pi)) for i in range(chebysh_precision)] 
chebyshev_mesh_y = [f(chebyshev_mesh_x[i]) for i in range (chebysh_precision)]

xgraph = [i * 0.01 for i in range(-100,100)]
ygraph = [f(xgraph[i]) for i in range (len(xgraph))]

reg_res = [lagrange(regular_mesh_x,regular_mesh_y,i) for i in xgraph]
cheb_res = [lagrange(chebyshev_mesh_x,chebyshev_mesh_y,i) for i in xgraph]
from matplotlib import pylab as plt
plt.style.use('fivethirtyeight')

plt.plot(xgraph,ygraph)
plt.plot(xgraph,reg_res)
plt.show()
```
