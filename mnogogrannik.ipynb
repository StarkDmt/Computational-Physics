```python
import numpy as np

#наша функция
def f(x,y,z):
    return (x)**2+(y)**2+(z)**2

#размерность пространства (число аргументов функции)    
N=3

#точность
eps=0.00001

#reflection
a=1.0

#contraction
b=0.2

#expansion
c=2.0

sigma=0.5

#генерация симплекса
simple = np.ndarray(shape = (N+1,N))
for i in range (N+1):
    for j in range (N):
        simple[i][j]=(-100)**(i)+(-100)**(i)*j*0.01

             
def cmp(x):
    return f(*x)
#отражение
def reflection (f,simplex):
    values = []
    
    #сортируем значения функции в симплексе
    for z in range (N+1):
        values.append( f(*simplex[z]) )

        
    simplex1=sorted(simplex,key=cmp)
    
    #центроид
   
    center = simplex1[0]
    z=1
    while (z<N):
        center = center + simplex1[z]
        z+=1
     
    center = center / N
   
    #отраженная точка
    r=center + a * ( center - simplex1[N] )
   
    return r
 
#расширение    
def expansion (f,simplex,xr):
    
    values = []
    
    #сортируем значения функции в симплексе
    for z in range (N+1):
        values.append( f(*simplex[z]) )
        

    simplex1=sorted(simplex,key=cmp)
    
    #центроид
    
    center = simplex1[0]
    z=1
    while (z<N):
        
        center = center + simplex1[z]
        z+=1
    center = center / N
        
    #отраженная точка
    e=center + c * ( xr - center )
    return e

#сжатие
def contraction (f,simplex):
    values = []
    
    #сортируем значения функции в симплексе
    for z in range (N+1):
        values.append( f(*simplex[z]) )
        
  
    simplex1=sorted(simplex,key=cmp)
    
    #центроид
    
    center = simplex1[0]
    z=1
    while (z<N):
        
        center = center + simplex1[z]
        z+=1
    center = center / N
        
    #отраженная точка
    p=center + b * ( simplex1[N] - center )
    
    return p
                
#модификация симплекса при отражении    
def step1(f,simplex,last):

    last+=1
    simplex=sorted(simplex,key=cmp)
  
    if(last>=100000):
        return simplex
    values = []
    
    for i in range (N+1):
        values.append( f(*simplex[i]) )
    values.sort()

    r=reflection(f,simplex)
    
    while ( ( f(*r) < values[N] ) and (values[0] <= f(*r)) ):
       
        simplex[N] = r
        last+=1
        if(last>=100000):
            return simplex
        r=reflection(f,simplex)
    if ( f(*r) < values[0] ):
        
        return step2(f,simplex,r,last)
    if (f(*r) >= values [N]):
        
        return step3(f,simplex,r,last)
    
    

#модификация симплекса при расширении
def step2(f,simplex,r,last):
    
    values = []
    for z in range (N+1):
        values.append( f(*simplex[z]) )
    values.sort()
    
    e=expansion(f,simplex,r)
    if(f(*e)<f(*r)):
        simplex[N] = e
    else:
        simplex[N] = r
    return step1(f,simplex,last)

#модификация симплекса при сжатии

def step3(f,simplex,r,last):
    values = []
    simplex1=simplex
    for z in range (N+1):
        values.append( f(*simplex1[z]) )
    values.sort()
    
    p=contraction(f,simplex1)
  
    if(f(*p) < values[N]):
        simplex1[N] = p
    else:
        for i in range (N):
            simplex[i+1]=simplex[0]+sigma*(simplex[i+1]-simplex[0])
    return step1(f,simplex1,last)
    
count=0
print(step1(f,simple,count)[1])
    


    
```
